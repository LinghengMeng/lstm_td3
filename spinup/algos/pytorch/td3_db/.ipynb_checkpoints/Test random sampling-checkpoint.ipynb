```python
class ToyReplayBuffer:
    """
    A simple FIFO experience replay buffer.
    """

    def __init__(self, size):
        self.exp_id_buf = np.zeros(size, dtype=np.float32)
        self.sampled_num_buf = np.zeros(size, dtype=np.float32)
        self.ptr, self.size, self.max_size = 0, 0, size

    def store(self, exp_id):
        self.exp_id_buf[self.ptr] = exp_id
        self.ptr = (self.ptr+1) % self.max_size
        self.size = min(self.size+1, self.max_size)

    def sample_batch(self, batch_size=32, sample_type = "pseudo_random"):
        # (1) pseudo_random  (2) genuine_random
        if sample_type == 'genuine_random':
#             sample_weights = (self.max_size - self.sampled_num_buf) / self.max_size
#             sample_weights = (self.size - self.sampled_num_buf) / self.size
#             sample_weights = 1 / (self.sampled_num_buf + 1e-1)
            
            sample_weights = 1 / (self.sampled_num_buf + 0.001)
            # sample_weights = np.exp(-self.sampled_num_buf)
            
            # Replacement
            batch_size = min(batch_size, self.size)
            
#             # With replacement
#             idxs = random.choices(self.exp_id_buf[:self.size], sample_weights[:self.size], k=batch_size)
            
            # Without replacement
            if batch_size>self.size:
                batch_size = self.size
            idxs = np.random.choice(self.exp_id_buf[:self.size], size=batch_size, replace=False, p=sample_weights[:self.size] / sample_weights[:self.size].sum())
            
        elif sample_type == "":
            pass
        elif sample_type == 'pseudo_random':
            idxs = np.random.randint(0, self.size, size=batch_size)

        # Increase sampled_num by 1
        for i in idxs:
            self.sampled_num_buf[int(i)] += 1

        return 0
```

```python
replay_size = 5000
```

## Pseudo_random sampling

```python
replay_buffer = ToyReplayBuffer(replay_size)
```

```python
batch_size = 100
sample_type = "pseudo_random"
```

```python
for i in range(replay_size):
    replay_buffer.store(i)
    replay_buffer.sample_batch(batch_size, sample_type)
```

## Genuine_random sampling

```python
genuine_replay_buffer = ToyReplayBuffer(replay_size)
```

```python
batch_size = 100
sample_type = "genuine_random"
```

```python
for i in range(replay_size):
    genuine_replay_buffer.store(i)
    genuine_replay_buffer.sample_batch(batch_size, sample_type)
```
